```python
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
titres = pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/a98cced2-babe-45b4-95dd-ce5336155dbc", sep=";")
```

```python
aides = pd.read_excel("https://www.data.gouv.fr/fr/datasets/r/480a55d3-b867-41f3-9af0-e1f3bff7ab1f")
```

```python
titres
```

```python
aides
```

```python
titres_aidés = titres[ titres.Titre.isin(aides["TABLEAU DES TITRES DE PRESSE AIDES EN 2016"]) ]
```

```python
titres_aidés
```

```python
titres_aidés = titres_aidés.join(aides.set_index('TABLEAU DES TITRES DE PRESSE AIDES EN 2016'), rsuffix=' - aides', on="Titre")
titres_aidés.rename(index=str, columns={"Unnamed: 8": "total"}, inplace=True)
titres_aidés
```

```python
titres_aidés_sum = titres_aidés.groupby("Editeur").sum()
```

```python
list(titres_aidés_sum)
```

```python
titres_aidés_sum.sort_values(by=['total'], ascending=False, inplace=True)
titres_aidés_sum[["Titre", "total"]]
```

```python
titres_aidés_count = titres_aidés.groupby("Editeur").count()
titres_aidés_count.rename(columns={"Titre": "titres"}, inplace=True)
titres_aidés_count[["titres"]]
```

```python
titres_aidés_sum[["total"]].join(titres_aidés_count[["titres"]])
```
