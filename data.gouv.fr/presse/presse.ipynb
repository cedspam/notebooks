```python
# data wrangling
import pandas as pd

# data visualisation
import altair as alt
alt.renderers.enable('notebook')
```

```python
# https://www.data.gouv.fr/fr/datasets/5bb834e89ce2e7693d2456a5/
df = pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/a98cced2-babe-45b4-95dd-ce5336155dbc", sep=";")
```

```python
df
```

```python
len(df)
```

```python
editeurs = df.groupby("Editeur").size()
editeurs = pd.DataFrame(editeurs).reset_index()
editeurs.rename(columns={ 0: 'titres' }, inplace=True)
editeurs.sort_values("titres", inplace=True, ascending=False)

editeurs
```

```python
editeurs.to_csv('editeurs.csv')
```
