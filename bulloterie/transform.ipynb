```python
import pandas as pd
```

```python
content = open("bulloterie.raw.txt", "r").read()
```

```python
content
```

```python
interests = content.split("\n\n")[1]

def map_line(line):    

    (interest, level, names) = line.split(" : ")
    
    return list(map(lambda x: (interest, level, x), names.split(" | ")))


def map_interests(txt):
    lines = txt.split("\n")

    records = [ map_line(l) for l in lines if l != "" ]
    
    return records


columns = ("intérêt", "niveau", "individu")
records = [ record for sl in map_interests(interests) for record in sl ]

df = pd.DataFrame.from_records(records, columns=columns)

df
```

```python
df.to_json("bulloterie.json", orient="records")
df.to_csv("bulloterie.csv", index=False)
```
